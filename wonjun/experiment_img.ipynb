```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Plotting 함수
- data 분포 확인
- min, max 확인
- mean, std 확인

```python
def plotting_data(data_csv):
    # 주어진 데이터
    x1 = len(data_csv)
    y1 = data_csv['mask_rle'].apply(lambda x: np.sum(list(map(int, x.split()[1:][::2])))/(1024*1024)).values

    # 그래프의 x축 범위
    x_range = np.linspace(0, 1, num = 3000)

    # x축 각 구간에 속하는 데이터 개수 계산
    hist, bin_edges = np.histogram(y1, bins=x_range)

    # 점의 크기 및 투명도 설정
    marker_sizes = np.where(hist > 0, 5, 0)
    alpha = 0.5

    # 그래프 그리기
    plt.scatter(x_range[:-1], hist, marker='o', s=marker_sizes, alpha=alpha)
    plt.xlabel('Y Ratio of building')
    plt.ylabel('Count')
    plt.title('Distribution of Y Values')
    plt.show()
    print("min ratio: ", min(y1), "max ratio: ", max(y1))
    print("mean", np.mean(y1), "std", np.std(y1))
```

```python
base = '../open (1)/'
train = pd.read_csv(f"{base}/train.csv")
```

```python
train[:1000].to_csv(f'{base}/zerotothousand.csv', index = False)
train[:500].to_csv(f'{base}/zeroto5hund.csv', index = False)
train[:20].to_csv(f'{base}/zeroto20.csv', index = False)
```

```python
ztt = pd.read_csv(f"{base}/zerotothousand.csv")
zt5 = pd.read_csv(f"{base}/zeroto5hund.csv")
zt20 = pd.read_csv(f"{base}/zeroto20.csv")
```

```python
plotting_data(train)
```

```python
plotting_data(ztt)
```
